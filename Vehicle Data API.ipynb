```python
## Import Libraries
import pandas as pd
import sqlalchemy as sa
import requests
```

```python
# 2006 Commercial Vehicle Survey: origin and destination
resource_id = '105d8297-7352-4313-ac3b-c6b4eac8ea6c'
limit = 1000
offset = 0
```

```python
# Define the API URL format with placeholders for variables
api_url = 'https://data.ontario.ca/api/3/action/datastore_search?resource_id={}&limit={}&offset={}'.format(resource_id,limit,offset)

print(api_url)
```

```python
# The 'api_response' variable now contains the server's response to the HTTP GET request.
# A status code of 200 indicates success (OK).
# A status code other than 200 may indicate an error or some other response.

api_response=requests.get(api_url)
api_response
```

```python
# The 'data' variable now holds the JSON data returned by the Ontario data portal API.
data=api_response.json()
data
```

```python
# The 'records' variable contains the list of records retrieved from the dataset.

data['result']['records']
```

```python
# Convert the list of records into a pandas DataFrame named 'Commercial_Vehicle_Survey'

Commercial_Vehicle_Survey=pd.DataFrame(data['result']['records'])
Commercial_Vehicle_Survey.tail()
```

```python
# Check for missing values (NaN or None) in the 'Commercial_Vehicle_Survey' DataFrame
Commercial_Vehicle_Survey.isnull()
```

```python
# Replace hyphens ('-') with missing values (pd.NA) in the 'Commercial_Vehicle_Survey' DataFrame
Commercial_Vehicle_Survey= Commercial_Vehicle_Survey.replace('-', pd.NA)
```

```python
# Assign the 'Commercial_Vehicle_Survey' DataFrame to a new variable 'df'
df=Commercial_Vehicle_Survey
```

```python
df
```

```python
df.info()
```

```python
# Clean the 'comodity weight (KG)' column: convert to numeric data type, remove commas, and coerce non-convertible values to missing values
df['comodity weight (KG)'] = pd.to_numeric(df['comodity weight (KG)'].str.replace(',', ''), errors='coerce')
# Clean the 'Commodity Value ($)' column: convert to numeric data type, remove commas, and coerce non-convertible values to missing values
df['Commodity Value ($)'] = pd.to_numeric(df['Commodity Value ($)'].str.replace(',', ''), errors='coerce')
```

```python
# Removing rows with missing values in the 'Daily trips' column
df = df.dropna(subset=['Daily trips'])
```

```python
# Feature: Total Value per Trip
df['Total Value per Trip'] = df['Daily trips'] * df['Commodity Value ($)']
```

```python
# Feature: Average Weight per Trip
df['Average Weight per Trip'] = df['comodity weight (KG)'] / df['Daily trips']
```

```python
mean_comodity_weight = df['comodity weight (KG)'].mean()
mean_commodity_value = df['Commodity Value ($)'].mean()
mean_total_value_per_trip = df['Total Value per Trip'].mean()
mean_average_weight_per_trip = df['Average Weight per Trip'].mean()
```

```python
df['comodity weight (KG)'].fillna(mean_comodity_weight, inplace=True)
df['Commodity Value ($)'].fillna(mean_commodity_value, inplace=True)
df['Total Value per Trip'].fillna(mean_total_value_per_trip, inplace=True)
df['Average Weight per Trip'].fillna(mean_average_weight_per_trip, inplace=True)

# Verify that there are no more missing values
print(df.isnull().sum())
```
